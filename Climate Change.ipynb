## Climate Change

```r
climate = read.csv("climate_change.csv")
train = subset(climate, Year <= 2006)
test = subset(climate, Year > 2006)
```

```r
rm1 = lm(Temp ~ MEI + CO2 + CH4 + N2O + CFC.11 + CFC.12 + TSI + Aerosols, data = train)
```

```r
summary(rm1)
```

If the p value is below 0.05, the variable is not an important factor in the model. In this model, we can see that 
* CH4 &
* N20 
have little effect on the temperature.

We also know that N20 and CFC.11 have -ive coefficients showing that they are strongly correlated with other variables in the data set

```r
cor(train)
```

```r
Taking out variables that are correlated to others having only unique ones remaining...
```

```r
rm2 = lm(Temp ~ MEI  + N2O + TSI + Aerosols, data = train)
```

```r
summary(rm2)
```

```r
rm1_simple = step(rm1)
```

```r
summary(rm1_simple)
```

```r
predict_temp = predict(rm1_simple, newdata = test)
SSE = sum((predict_temp - test$Temp)^2)
SST = sum((mean(train$Temp) - test$Temp)^2)
R2 = (1 - (SSE/SST))
```
